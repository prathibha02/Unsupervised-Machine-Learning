```python
# Initial imports
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = './Resources/myopia_cleaned.csv'
```

```python
df = pd.read_csv(file_path)
```

```python
# Initialize PCA model
pca = PCA(n_components=0.90)

# Get two principal components for the data.
myopia_pca = pca.fit_transform(df)
```

```python
# Transform PCA data to a DataFrame
df_myopia_pca = pd.DataFrame(
    data=myopia_pca,columns=["principal component 1", "principal component 2","principal component 3"]
)
df_myopia_pca.head()
```

```python
#Fetch the explained variance
pca.explained_variance_ratio_
```

```python
tsne = TSNE(learning_rate=340)
tsne_features = tsne.fit_transform(df_myopia_pca)
```

```python
# Check that dimensions have been reduced to two
tsne_features.shape
```

```python
# Results colored by target value
plt.scatter(tsne_features[:,0], tsne_features[:,1], c=tsne_features[:,1])
plt.show()
```
